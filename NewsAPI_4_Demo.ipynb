```python
import spotipy
import webbrowser
import json
import urllib.request
import geocoder
from nltk.corpus import stopwords
import spotipy.util as util
import random
```

```python
#I assigned my variable "key" to the API key from the news
#API
with open("My_API_Key.txt", "r") as file:
    key = file.read()
```

```python
#Since "key" has been assigned a value of the obtained API key
#I can substitute that into the necessary part of the URL
location = ""
#https://newsapi.org/v2/everything?q=Apple&from=2022-11-30&sortBy=popularity&apiKey=API_KEY
url = ('https://newsapi.org/v2/everything?q=Apple&from=2022-11-30&sortBy=popularity&apiKey=' + key)
```

```python
print(url)
```

```python
request = urllib.request.Request(url)
response = urllib.request.urlopen(request)
```

```python
type(response)
```

```python
#converting the http above to a python dictionary
headlines = json.loads(response.read())
```

```python
#I have printed the headlines to make sure this works
print(headlines)
```

```python
#I am pulling out all the 'unique' words from tje headlines
unique_words = set()
for headline in headlines['articles']:
    for word in headline['title'].lower().split():
        if word not in stopwords.words('english') and word.isalpha():
            unique_words.add(word)
            
unique_words
```
